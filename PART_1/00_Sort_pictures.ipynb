```python
import os

def delete_non_pictures(directory):
    if not os.path.isdir(directory):
        print(f"Error: The directory {directory} does not exist.")
        return

    # Define a set of valid picture extensions
    valid_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp'}
    # Count files deleted
    deleted_files = 0

    # List files in the directory
    files = os.listdir(directory)
    for file in files:
        # Check the extension of each file
        if not any(file.lower().endswith(ext) for ext in valid_extensions):
            # Build the full file path
            file_path = os.path.join(directory, file)
            # Delete the file
            try:
                os.remove(file_path)
                deleted_files += 1
                print(f"Deleted: {file_path}")
            except Exception as e:
                print(f"Failed to delete {file_path}: {e}")

    print(f"Total deleted files: {deleted_files}")

# Get user input for the directory path
directory_path = input("Enter the directory path: ")
delete_non_pictures(directory_path)
```

```python
import os
from PIL import Image

def resize_images_in_directory(source_directory, target_directory, size=(224, 224)):
    # Check if the source directory exists
    if not os.path.isdir(source_directory):
        print("The specified source directory does not exist.")
        return

    # Create target directory if it does not exist
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)
        print(f"Created directory {target_directory}")

    # Iterate over all files in the source directory
    for filename in os.listdir(source_directory):
        source_filepath = os.path.join(source_directory, filename)

        # Check if it's a file
        if os.path.isfile(source_filepath):
            try:
                # Open the image
                with Image.open(source_filepath) as img:
                    # Resize the image using LANCZOS resampling
                    img = img.resize(size, Image.Resampling.LANCZOS)
                    
                    # Define the path for the resized image in the target directory
                    target_filepath = os.path.join(target_directory, filename)
                    
                    # Save the resized image to the target directory
                    img.save(target_filepath)
                    print(f"Resized and saved {filename} in {target_directory}")
            except IOError:
                print(f"Cannot resize {filename}, it might not be an image.")

# Replace 'path_to_source_directory' with the path to your directory containing images
# Replace 'path_to_target_directory' with the path where you want to save resized images
resize_images_in_directory(r'Orignal Datasets\NOT Auguste_Original', r'224_Datasets\NOT Auguste_224')
```
