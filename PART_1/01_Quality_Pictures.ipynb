```python
import os
from PIL import Image
import hashlib
from tqdm import tqdm

def hash_file(filename):
    """Generate MD5 hash for a file."""
    hasher = hashlib.md5()
    with open(filename, 'rb') as f:
        buf = f.read()
        hasher.update(buf)
    return hasher.hexdigest()

def remove_duplicates(directory):
    hashes = {}
    files_to_remove = []

    # Walk through all images and calculate their hashes
    for filename in tqdm(os.listdir(directory)):
        file_path = os.path.join(directory, filename)
        if filename.lower().endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif')):
            image_hash = hash_file(file_path)
            if image_hash in hashes:
                files_to_remove.append(file_path)
            else:
                hashes[image_hash] = file_path

    # Remove duplicate images
    for file_path in files_to_remove:
        os.remove(file_path)
        print(f"Removed duplicate file: {file_path}")

    print(f"Total duplicates removed: {len(files_to_remove)}")

# Replace 'path_to_directory' with the path to your directory containing images
remove_duplicates(r'224_Datasets\NOT Auguste_224')
```
